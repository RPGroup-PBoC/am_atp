```python
file_path = '../../../data/aster/2023-03-06_NCD(3-3-23)/'
recy_folder = '2023-03-06_NCD(3-3-23)_Asters_Recycling_1/Pos0/'
none_folder = '2023-03-06_NCD(3-3-23)_Asters_2/Pos0/'

recyYFP = pd.read_csv(file_path+recy_folder+'YFP.csv')
recyCFP = pd.read_csv(file_path+recy_folder+'CFP.csv')
noneYFP = pd.read_csv(file_path+none_folder+'YFP.csv')
noneCFP = pd.read_csv(file_path+none_folder+'CFP.csv')
```

```python
fig,ax = plt.subplots(figsize = (8,4))
ax.plot(recyYFP['X']/3, recyYFP['Y'], color = 'goldenrod', linestyle = '--', label = 'Recycling YFP')
ax.plot(recyCFP['X']/3, recyCFP['Y'], color = 'c', linestyle = '--', label = 'Recycling CFP')
ax.plot(noneYFP['X']/3, noneYFP['Y'], color = 'goldenrod', label = 'No Recycling YFP')
ax.plot(noneCFP['X']/3, noneCFP['Y'], color = 'c', label = 'No Recycling CFP')
plt.legend()
```

```python
fig, ax = plt.subplots(2,2, figsize = (12,8))
ax[0,0].plot(recyYFP['X']*20/60, recyYFP['Y'], color = 'goldenrod')
ax[1,0].plot(recyCFP['X']*20/60, recyCFP['Y'], color = 'c')
ax[0,1].plot(noneYFP['X']*20/60, noneYFP['Y'], color = 'goldenrod')
ax[1,1].plot(noneCFP['X']*20/60, noneCFP['Y'], color = 'c')

ax[0,0].set_ylabel(r'YFP', fontsize=25)
ax[1,0].set_ylabel(r'CFP', fontsize=25)
ax[1,0].set_xlabel('Time (minutes)', fontsize = 25)
ax[1,1].set_xlabel('Time (minutes)', fontsize = 25)
ax[0,0].set_title('With Recycling', fontsize = 30)
ax[0,1].set_title('Without Recycling', fontsize =30)

ax[0,0].tick_params('both', which = 'both', width=2, labelsize=20)
ax[1,0].tick_params('both', which = 'both', width=2, labelsize=20)
ax[0,1].tick_params('both', which = 'both', width=2, labelsize=20)
ax[1,1].tick_params('both', which = 'both', width=2, labelsize=20)

plt.tight_layout()
```
